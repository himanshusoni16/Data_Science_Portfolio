```python
## now make the Date column as index of the dateset
asp = pd.read_csv("AAPL _stock_dataset.csv",parse_dates=["Date"],index_col= "Date")
asp.head()
```

```python
## TO varifie by 
asp.index
## we see that this index is type datetime index 
```

```python
## why we change data column as index?
## i want to retrieve only jan 2017 stock prices so,

asp["2017-01"] ## it return only specifice month data 
```

```python
## Find the averge price of apple stock in jan 2017
asp["2017-01"].Close.mean()
```

```python
## what is the averge monthly price of  apple stock dataset from( jun 2017- jun 2018)
asp["2017-01":"2018-01"].Close.resample("M").mean()
```

```python
## plot the is the average monthly price of  apple stock dataset from( jun 2017- jun 2018)
asp["2017-01":"2018-01"].Close.resample("M").mean().plot(kind="bar")

##  we fine the monthly average price increasing 
```

```python
asp.columns
```

```python
## rename the "Adj Close" column 
asp.rename(columns= {'Adj Close' : "Adj_close"}, inplace = True )
```

```python
# plot the is the average monthly price of  apple stock dataset from( jun 2017- jun 2018 of Adj_close 
asp["2017-01":"2018-01"].Adj_close.resample("M").mean().plot(kind="bar")
```

```python
asp.describe() ## give basic description of the dataset just like ( mean , minimum , maximum, count )
```

```python
# What is the change in stock's price over time?
asp['Adj_close'].plot(legend = True, figsize=(12,5))
```

```python
##Using pandas once more to plot the total volume being traded over time
asp['Volume'].plot(legend=True,figsize=(12,5))
```

```python
# plot monthly frequncy Ajd_close price chart 
%matplotlib inline
asp.Adj_close.resample("M").mean().plot(legend = True, figsize=(12,5))
```

```python
# plot quaterly  frequncy Ajd_close avrage price chart 
%matplotlib inline
asp.Adj_close.resample("Q").mean().plot(legend = True, figsize=(12,5))
```

```python
asp[['Open','High','Low','Close','Adj_close']].plot(legend = True, figsize=(12,5))
```

```python
# What was the daily return average of a stock?
# we use the {pct_change()}--Computes the percentage change from the immediately previous row by default.
# This is useful in comparing the percentage of change in a time series of elements.
# make a new cloumn in the dataset { "Daily_Return"}

asp["Daily_Return"] = asp["Adj_close"].pct_change()
asp.tail()
```

```python
## chart plot of the Daily return
asp['Daily_Return'].plot(legend = True, figsize=(12,5))
```

```python
# histogram of daily_return
asp['Daily_Return'].plot(kind='hist',figsize=(12,5))
### Positive daily returns seem to be more frequent than negative returns
```

```python
#Let's use a pairplot to visualize all in one view.
sns.pairplot(asp[['Open','High','Low','Close','Adj_close']])
```
